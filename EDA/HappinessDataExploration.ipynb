Not the strongest correlations we have across our variables. Let's take a quick look at the outliers for corruption.

```python
from scipy.stats import chi2
from matplotlib import patches
```

```python
cov_df = cleaned_df[['Happiness score', 'Perceptions of corruption']]
cov_df = cov_df.to_numpy()

#calculating covariance matrix
covariance = np.cov(cov_df, rowvar=False)

#calculating covariance matrix power of -1
covariance_pm1 = np.linalg.matrix_power(covariance, -1)

#Center point
centerpoint = np.mean(cov_df, axis=0)
print('centerpoint', centerpoint)
```

```python
distances = []
for i, val in enumerate(cov_df):
  p1 = val
  p2 = centerpoint
  distance = (p1-p2).T.dot(covariance_pm1).dot(p1-p2)
  distances.append(distance)

distances = np.array(distances)
cutoff = chi2.ppf(0.95, cov_df.shape[1])
outlierIndexes = np.where(distances > cutoff )

print('--- Index of Outliers ---')
print(outlierIndexes)
print('--- Obversations found as outliers ---')
print(cov_df[distances > cutoff , :])
```

```python
print('The identified outliers for happiness score and perceptions of corruption correlation:')
corruption_df = cleaned_df.loc[outlierIndexes]
corruption_df
```

## Plotting outlier countries for corruption:

```python
f, ax = plt.subplots(figsize=(30, 15))
sns.set_color_codes("pastel")
sns.barplot(x='Country', y='Happiness score', data=corruption_df,
            label='Happiness score', color='#F3B190')

sns.set_color_codes("muted")
sns.barplot(x='Country', y='Perceptions of corruption', data=corruption_df,
        label='Perceptions of corruption', color='#EB7A41')

ax.legend(ncol=2, loc="upper right", frameon=True)
ax.set(xlim=(0, 24), ylabel="",
    xlabel="Happiness Score")
f.suptitle('Comparing Happiness Score and Perceptions of Corruption')
sns.despine(left=True);
```
